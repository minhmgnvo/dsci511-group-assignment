__3a.__ (5 points) Create a function called `indices` that takes a string as input, lowercases the string, and outputs a dictionary of indices indicating where each letter of the alphabet first occurs in the string. Test the function by printing the results for examples `x` and `y` provided above. (Note: Your function should `break` and `return` the list as soon as all `26` characters are found.)

```python
import string

def indices(text):
    alphabet = set(string.ascii_lowercase)
    index = {}
    arr = list(text.lower())
    for x in arr:
        if len(index) == 26:
            break
        if x in alphabet and x not in index:
            index[x] = arr.index(x)
    sorted_d = dict(sorted(index.items()))
    return sorted_d

print(indices(x))
print(indices(y))
```

__3b.__ (3 points) Create a function called `verify` that takes a string as input and uses the output of `indices` to determine if the string is a pangram. The function should return `True` if the string is a pangram and `False` otherwise. Test your function on examples `x` and `y`.

```python
def verify(text):
    alphabet =  set(string.ascii_lowercase)
    word_count = {letter: 0 for letter in alphabet}
    arr = list(text.lower())
    
    for x in arr:
        if x in alphabet:
            word_count[x] = 1

    if 0 in word_count.values():
        return "False"
    else: 
        return "True"

print("x is a pangram:", verify(x))
print("y is a pangram:", verify(y))
```

__3c.__ (2 points) Create a function called `tiny_verify()` that determines whether a string is a pangram in a single line of code without using `indices`. Test the function by printing the results for examples `x` and `y` provided above. (Hint: Use a list comprehension.)

```python
def tiny_verify(text):
    return all(letter in text.lower() for letter in string.ascii_lowercase)

print("x is a pangram:", tiny_verify(x))
print("y is a pangram:", tiny_verify(y))
```

__3d.__ (5 points) Create a function called `missing_letters` that takes a string as input, lowercases the string, and returns a list of letters that do not appear in the string. Test the function by printing the results for examples `x` and `y` provided above.

```python
from string import ascii_lowercase as ascii_letters

def missing_letters(text):
    lower = text.lower()
    not_in = []
    for x in ascii_letters:
        if x not in lower:
            not_in.append(x)
    return not_in

print("Letters missing from x:", missing_letters(x))
print("Letters missing from y:", missing_letters(y))
```

```python
test_text = """
Lorem ipsum dolor sit amet, consectetur adipiscing elit, sed do eiusmod tempor
incididunt ut labore et dolore magna aliqua. Ut enim ad minim veniam, quis
nostrud exercitation ullamco laboris nisi ut aliquip ex ea commodo consequat.
Duis aute irure dolor in reprehenderit in voluptate velit esse cillum dolore eu
fugiat nulla pariatur. Excepteur sint occaecat cupidatat non proident, sunt in
culpa qui officia deserunt mollit anim id est laborum.
"""

def sentokenize(text, delims = [".", "!", "?"]):
  if not delims:
    return [text.strip()] if text.strip() else []
  
  sentences = []
  for subst in text.split(delims[0]):
    sentences.extend(sentokenize(subst, delims[1:]))
  return sentences

# print(sentokenize(test_text))
for sentence in sentokenize(test_text):
  print(sentence)
  print("----")
```
